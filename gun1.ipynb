## Webcam ile kayıt almak

```python
import cv2
import datetime

# 1. Kamerayı başlatma
cap = cv2.VideoCapture(0)  # 0, varsayılan kamera cihazını temsil eder

# 2. Video yazıcı nesnesi oluşturma
# Codec formatı belirleme ('XVID') ve VideoWriter nesnesini ayarlama
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # 4-byte codec tanımı
out = cv2.VideoWriter('output3.mp4', fourcc, 20.0, (1000, 1000))  # Video yazıcı oluşturma

# 3. Genişlik ve yükseklik ayarlama (isteğe bağlı)
# cap.set(cv2.CAP_PROP_FRAME_WIDTH, 300)  # Genişliği ayarlama
# cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 150)  # Yüksekliği ayarlama

# 4. Döngü: Kamera açık olduğu sürece işlem yap
while cap.isOpened():  # Kamera düzgün bir şekilde açılmış mı kontrol et
    success, frame = cap.read()  # Kameradan bir kare al
    if not success:
        break  # Kare alınamazsa döngüden çık

    frame = cv2.flip(frame, 1)  # Görüntüyü yatay eksende (ayna gibi) çevirme

    # 5. Görüntü üzerine tarih ve saat ekleme
    datet = str(datetime.datetime.now())  # Güncel tarih ve saati al
    cv2.putText(frame, datet, (180, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 1)  # Tarih ve saat ekleme

    # 6. Kareyi video dosyasına yazdırma
    out.write(frame)  # Her bir kareyi video dosyasına ekle

    # 7. Görüntüyü gösterme
    cv2.imshow("webCam", frame)  # Kameradan alınan kareyi ekranda göster

    # 8. 'q' tuşuna basıldığında döngüyü sonlandırma
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# 9. Kaynakları serbest bırakma
cap.release()  # Kamerayı serbest bırak
out.release()  # Video dosyasını kapat
cv2.destroyAllWindows()  # Tüm OpenCV pencerelerini kapat
```

# Resimleri Birleştirme

```python
import cv2

#resimleri okuma
image1=cv2.imread("car.png")
image2=cv2.imread("car2.jpg")

#yan yana birleştirme
if image1.shape[0] != image2.shape[0]:  # Yükseklikler eşit değilse
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))  # İkinci görseli birinciye göre yeniden boyutlandırıyoruz

horizontal = cv2.hconcat([image1, image2])

#üst üste birleştirme
if image1.shape[1] != image2.shape[1]:  # Genişlikler eşit değilse
    image2= cv2.resize(image2, (image1.shape[1], image2.shape[0]))  # İkinci görseli birinciyle aynı genişlikte olacak şekilde yeniden boyutlandır

vertical = cv2.vconcat([image1, image2])

#yaptığımız görselleri gösterme
cv2.imshow("Yan Yana Birleştirme", horizontal)
cv2.imshow("Üst Üste Birleştirme", vertical)

#yaptığımız görselleri kaydetme
cv2.imwrite("Yan Yana Birlestirme.png", horizontal)
cv2.imwrite("Ust Uste Birlestirme.png", vertical)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blurlama

**Bulanıklaştırmanın avantajları:**

>Gürültü gidermede yardımcı olur. Gürültü yüksek geçiş sinyali olarak kabul edildiğinden, düşük geçiş filtresi çekirdeğinin uygulanmasıyla gürültüyü kısıtlarız.


>Görüntünün pürüzsüzleşmesine yardımcı olur.

>Düşük yoğunluklu kenarlar kaldırılır.


>Gerektiğinde ayrıntıları gizlemeye yardımcı olur.

**Önemli bulanıklaştırma türleri:**
>**Gaussian Blurring:** Gaussian blur, bir görüntünün Gaussian fonksiyonu ile bulanıklaştırılmasının sonucudur. Grafik yazılımlarında yaygın olarak kullanılan bir efekttir, genellikle görüntü gürültüsünü ve ayrıntıları azaltmak için kullanılır. Ayrıca makine öğrenimi veya derin öğrenme modellerimizi uygulamadan önce bir ön işleme aşaması olarak kullanılır. (kenarları yumuşatır)

>**Median Blur:** Median Filtresi, genellikle bir görüntü veya sinyalden gürültüyü gidermek için kullanılan doğrusal olmayan bir dijital filtreleme tekniğidir. Median filtreleme, belirli koşullar altında gürültüyü giderirken kenarları koruduğu için dijital görüntü işlemede çok yaygın olarak kullanılır. Tuz ve biber gürültüsünü gidermek için en iyi algoritmalardan biridir.

>**Bilateral Blur:** İkili filtre, görüntüler için doğrusal olmayan, kenar koruyan ve gürültü azaltan bir yumuşatma filtresidir. Her pikselin yoğunluğunu yakındaki piksellerden gelen yoğunluk değerlerinin ağırlıklı ortalamasıyla değiştirir. Bu ağırlık bir Gauss dağılımına dayanabilir. Böylece, keskin kenarlar korunurken zayıf olanlar atılır.

```python
import cv2 
import numpy as np 

image = cv2.imread('car2.jpg') 
image=cv2.resize(image,(400,500))
# Gaussian Blur 
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 

# Median Blur 
median = cv2.medianBlur(image, 5) 

# Bilateral Blur 
bilateral = cv2.bilateralFilter(image, 9, 75, 75) 


combined_image = cv2.hconcat([image, Gaussian, median, bilateral]) #çıktılarımızı birleştirmek


#cv2.imshow('Original Image', image) 
#cv2.imshow('Gaussian Blurring', Gaussian)
#cv2.imshow('Median Blurring', median) 
#cv2.imshow('Bilateral Blurring', bilateral) 
cv2.imshow('tumu', combined_image) 

#tümü olan görseli indirmek
cv2.imwrite("tum_blur.png", combined_image)

cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

# Görüntü Eşikleme(Threshold)

<code>cv2.threshold(source, thresholdValue, maxVal, thresholdingTechnique) </code>

-> **source:** Input Image array (must be in Grayscale). <br>
-> **thresholdValue:** Value of Threshold below and above which pixel values will change accordingly. <br>
-> **maxVal:** Maximum value that can be assigned to a pixel. <br>
-> **thresholdingTechnique:** The type of thresholding to be applied. <br>


### Eşikleme Mantığı Nedir?
Bir görüntüdeki her pikselin bir parlaklık veya yoğunluk değeri vardır. Eşikleme işlemi, bu piksel değerlerinin belirli bir eşik değerine göre karşılaştırılmasını sağlar. Bir pikselin değeri eşik değerinden küçükse veya büyükse, o piksel bir değere atanır (örneğin 0 ya da 255). Bu şekilde görüntüdeki belirli alanlar daha net bir şekilde ayırt edilir.

## Basit Eşikleme Teknikleri
Şimdi, OpenCV'deki bazı temel eşikleme yöntemlerini inceleyelim.

### 1. THRESH_BINARY:
Eşikleme işleminde en temel tekniktir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 255 (beyaz) olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, bu piksel değeri 0 (siyah) olarak atanır.<br>
Örnek: Bir eşik değeri 127 verilmişse, 128 ve üstü tüm pikseller beyaz yapılır, 126 ve altındaki tüm pikseller siyah yapılır.<br>

### 2. THRESH_BINARY_INV:
THRESH_BINARY'nin tam tersidir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 0 (siyah) olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, bu piksel değeri 255 (beyaz) olarak atanır.<br>
Örnek: 127 eşik değeriyle çalışıyorsak, 128 ve üstü pikseller siyah yapılırken, 126 ve altı tüm pikseller beyaz yapılır.<br>

### 3. THRESH_TRUNC:
Piksel değeri eşik değerinden büyükse, bu piksel değeri eşik değerine ayarlanır.<br>
Piksel değeri eşik değerinden küçükse, olduğu gibi bırakılır.<br>
Örnek: 127 eşik değeriyle, 130 değerinde olan bir piksel 127’ye çekilirken, 100 değerinde olan bir piksel aynı kalır.<br>

### 4. THRESH_TOZERO:
Piksel değeri eşik değerinden küçükse, bu piksel değeri 0 olarak atanır (siyah yapılır).<br>
Piksel değeri eşik değerinden büyükse, pikselin orijinal değeri korunur.<br>
Örnek: 127 eşik değeriyle, 126 ve altı değerlerdeki pikseller siyah yapılırken, 128 ve üstü değerler orijinal haliyle bırakılır.<br>

### 5. THRESH_TOZERO_INV:
THRESH_TOZERO işleminin tersidir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 0 olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, orijinal piksel değeri korunur.<br>
Örnek: 127 eşik değeriyle, 126 ve altı değerlerdeki pikseller orijinal haliyle bırakılırken, 128 ve üstü pikseller siyah yapılır.<br>

### Eşikleme Nerelerde Kullanılır?
Eşikleme genellikle nesne tespiti, kontur bulma veya nesne segmentasyonu gibi işlemlerde kullanılır. Örneğin, bir belge tarayıp sadece metinleri ön plana çıkarmak istediğinizde, eşikleme ile siyah-beyaz bir görüntü elde edebilirsiniz.

# Simple Thresholding


```python
import cv2  
import numpy as np  

# 1. Giriş görüntüsünü okuma
# 'apple1.jpg' adlı resmi yüklüyoruz
image = cv2.imread('apple1.jpg')  

# 2. Gri tonlamalı hale çevirme
# Görüntüyü siyah-beyaz işleme için gri tonlamalı hale getiriyoruz
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  

# 3. Eşikleme tekniklerini uygulama
# 120 eşik değerinin üzerinde olan pikseller 255 (beyaz) olarak ayarlanacak

# a) Binary Thresholding: 120'den büyük pikseller beyaz (255) olacak, diğerleri siyah (0) olacak
ret, thresh1 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY) 

# b) Binary Inverted Thresholding: 120'den küçük pikseller beyaz (255) olacak, diğerleri siyah (0) olacak
ret, thresh2 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY_INV) 

# c) Truncated Thresholding: 120'den büyük pikseller 120'ye indirgenecek, diğerleri olduğu gibi kalacak
ret, thresh3 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TRUNC) 

# d) To Zero Thresholding: 120'den küçük pikseller sıfırlanacak, diğerleri olduğu gibi kalacak
ret, thresh4 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TOZERO) 

# e) To Zero Inverted Thresholding: 120'den büyük pikseller sıfırlanacak, diğerleri olduğu gibi kalacak
ret, thresh5 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TOZERO_INV)  

# 4. Eşikleme sonucu görüntüleri gösterme
# Farklı eşikleme yöntemlerinin çıktıları her biri için ayrı pencere ile gösterilir
cv2.imshow('Binary Threshold', thresh1)  # Binary thresholding
cv2.imshow('Binary Threshold Inverted', thresh2)  # Inverted binary thresholding
cv2.imshow('Truncated Threshold', thresh3)  # Truncated thresholding
cv2.imshow('Set to 0', thresh4)  # To zero thresholding
cv2.imshow('Set to 0 Inverted', thresh5)  # To zero inverted thresholding

# 5. Görüntüleri kapama
# Kullanıcı bir tuşa bastığında tüm pencereleri kapatıyoruz
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Adaptive Thresholding
Basit Eşiklemede, her zaman sabit kalan global bir eşik değeri kullanılmıştır. Bu nedenle, sabit bir eşik değeri farklı alanlardaki değişken aydınlatma koşullarında yardımcı olmayacaktır. Uyarlanabilir eşikleme, eşik değerinin daha küçük bölgeler için hesaplandığı yöntemdir. Bu, aydınlatmadaki değişime göre farklı bölgeler için farklı eşik değerlerine yol açar. Bunun için cv2.adaptiveThreshold kullanıyoruz.

<code>cv2.adaptiveThreshold(source, maxVal, adaptiveMethod, thresholdType, blocksize, constant)</code>


**Parametreler:**

-> **source:** Giriş Görüntü dizisi (Tek kanallı, 8 bit veya kayan noktalı)

-> **maxVal:** Bir piksele atanabilecek maksimum değer.

-> **adaptiveMethod:** Adaptif yöntem eşik değerinin nasıl hesaplanacağına karar verir.


 **cv2.ADAPTIVE_THRESH_MEAN_C:** Eşik Değeri = (Komşuluk alanı değerlerinin ortalaması - sabit değer). Başka bir deyişle, bir noktanın blockSize×blockSize komşuluğunun ortalaması eksi sabittir.


**cv2.ADAPTIVE_THRESH_GAUSSIAN_C:** Eşik Değeri = (Komşuluk değerlerinin Gauss ağırlıklı toplamı - sabit değer). Başka bir deyişle, bir noktanın blockSize×blockSize komşuluğunun ağırlıklı toplamı eksi sabittir.

-> **thresholdType:** Uygulanacak eşikleme türü.

-> **blockSize:** Eşik değerini hesaplamak için kullanılan piksel komşuluğunun boyutu.

-> **constant:** Komşuluk piksellerinin ortalamasından veya ağırlıklı toplamından çıkarılan sabit bir değer.

```python
import cv2  
import numpy as np  
  
# 1. Giriş görüntüsünü okuma
# 'apple1.jpg' adlı resmi yüklüyoruz
image = cv2.imread('apple1.jpg')  
  
# 2. Gri tonlamalı hale çevirme
# Görüntüyü siyah-beyaz işleme için gri tonlamalı hale getiriyoruz
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  

# 3. Adaptif eşikleme uygulama (Mean)
# Görüntü üzerinde adaptif eşikleme (Mean) yöntemiyle siyah-beyaz hale getirme
# Bu yöntem, her piksel için çevresindeki piksellere bağlı olarak bir eşik değeri hesaplar
thresh1 = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 

# 4. Adaptif eşikleme uygulama (Gaussian)
# Görüntü üzerinde adaptif eşikleme (Gaussian) yöntemiyle siyah-beyaz hale getirme
# Bu yöntem de çevresindeki piksellere bağlı olarak, ancak daha farklı bir hesaplama ile eşik değeri belirler
thresh2 = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
  
# 5. Eşikleme sonucu görüntüleri gösterme
# 'Adaptive Mean' ve 'Adaptive Gaussian' tekniklerinin sonuçlarını görselleştiriyoruz
cv2.imshow('Adaptive Mean', thresh1)  # Mean yöntemiyle eşiklenmiş görüntü
cv2.imshow('Adaptive Gaussian', thresh2)  # Gaussian yöntemiyle eşiklenmiş görüntü

# 6. Görüntüleri kapatma
# Kullanıcı bir tuşa bastığında tüm pencereleri kapatıyoruz
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Otsu Threshold

<code>cv2.threshold(source, thresholdValue, maxVal, thresholdingTechnique)</code>


**Parameters:** 

-> **source:** Input Image array (must be in Grayscale).

-> **thresholdValue:** Value of Threshold below and above which pixel values will change accordingly.

-> **maxVal:** Maximum value that can be assigned to a pixel.

-> **thresholdingTechnique:** The type of thresholding to be applied.

```python
import cv2          
import numpy as np     
  
# 1. Görüntüyü yükleme
# Belirtilen dosya yolundaki görüntüyü yüklüyoruz
image = cv2.imread('apple1.jpg') 
  
# 2. Görüntüyü gri tonlamaya çevirme
# cv2.cvtColor() ile görüntüyü gri tonlamaya çeviriyoruz
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
  
# 3. Otsu eşikleme uygulama
# cv2.threshold() kullanarak gri tonlamalı görüntüye Otsu eşikleme yapıyoruz
# THRESH_OTSU, otomatik olarak bir eşik değeri belirliyor
ret, thresh1 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)      

# 4. Görüntüleri gösterme
# Orijinal görüntüyü ve Otsu eşikleme sonrası oluşan görüntüyü ekranda gösteriyoruz
cv2.imshow('Original Image', image)  # Orijinal görüntü
cv2.imshow('Otsu Threshold', thresh1)  # Eşikleme sonrası görüntü

# 5. Kullanıcı bir tuşa basana kadar bekliyoruz
cv2.waitKey(0)

# 6. Tüm pencereleri kapatma
# Açık olan tüm pencereleri kapatıyoruz
cv2.destroyAllWindows()
```

# Morfolojik Operasyonlar

## Erosion(aşındırma) ve Dilation(genişletme)

```python
import cv2 
import numpy as np 

# 1. Giriş görüntüsünü okuma
# Giriş olarak verilen resmi yüklüyoruz
img = cv2.imread('ataturk.png')

# 2. Gri tonlamalı hale çevirme
# Görüntüyü siyah-beyaz işlem için gri tonlamalı hale çeviriyoruz
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Eşikleme işlemi uygulama
# cv2.threshold() ile görüntüyü siyah ve beyaz olmak üzere iki renge ayırıyoruz
thresh, BW_im = cv2.threshold(gray_img, 150, 255, cv2.THRESH_BINARY)

# 4. Çekirdek matrisi tanımlama
# Görüntü işleme işlemleri için 5x5 boyutunda bir çekirdek matrisi oluşturuyoruz
kernel = np.ones((5, 5), np.uint8)

# 5. Erozyon uygulama
# Görüntünün belirli bölgelerini küçültmek için erozyon işlemi yapıyoruz
img_erosion = cv2.erode(BW_im, kernel, iterations=1)

# 6. Dilasyon uygulama
# Görüntünün belirli bölgelerini genişletmek için dilasyon işlemi yapıyoruz
img_dilated = cv2.dilate(BW_im, kernel, iterations=1)

# 7. Görüntüleri gösterme
# Orijinal görüntüyü, siyah-beyaz hale getirilmiş görüntüyü ve işlem sonrası görüntüleri gösteriyoruz
cv2.imshow('Orijinal Görsel', img)  # Orijinal görüntü
cv2.imshow('Siyah-Beyaz', BW_im)  # Siyah-beyaz görüntü
cv2.imshow('Erozyon Uygulanmış', img_erosion)  # Erozyon sonrası görüntü
cv2.imshow('Dilasyon Uygulanmış', img_dilated)  # Dilasyon sonrası görüntü

# 8. Görüntüleri kaydetme
# Erozyon ve dilasyon uygulanmış görüntüleri dosyaya kaydediyoruz
cv2.imwrite("erozyonlu.png", img_erosion)
cv2.imwrite("dilasyonlu.png", img_dilated)

# 9. Görüntü pencerelerini kapatma
# Kullanıcı bir tuşa basınca tüm pencereleri kapatıyoruz
cv2.waitKey(0)
cv2.destroyAllWindows()
```
