# Resimleri Birleştirme

```python
import cv2

#resimleri okuma
image1=cv2.imread("car.png")
image2=cv2.imread("car2.jpg")

#yan yana birleştirme
if image1.shape[0] != image2.shape[0]:  # Yükseklikler eşit değilse
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))  # İkinci görseli birinciye göre yeniden boyutlandırıyoruz

horizontal = cv2.hconcat([image1, image2])

#üst üste birleştirme
if image1.shape[1] != image2.shape[1]:  # Genişlikler eşit değilse
    image2= cv2.resize(image2, (image1.shape[1], image2.shape[0]))  # İkinci görseli birinciyle aynı genişlikte olacak şekilde yeniden boyutlandır

vertical = cv2.vconcat([image1, image2])

#yaptığımız görselleri gösterme
cv2.imshow("Yan Yana Birleştirme", horizontal)
cv2.imshow("Üst Üste Birleştirme", vertical)

#yaptığımız görselleri kaydetme
cv2.imwrite("Yan Yana Birlestirme.png", horizontal)
cv2.imwrite("Ust Uste Birlestirme.png", vertical)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blurlama

**Bulanıklaştırmanın avantajları:**

>Gürültü gidermede yardımcı olur. Gürültü yüksek geçiş sinyali olarak kabul edildiğinden, düşük geçiş filtresi çekirdeğinin uygulanmasıyla gürültüyü kısıtlarız.


>Görüntünün pürüzsüzleşmesine yardımcı olur.

>Düşük yoğunluklu kenarlar kaldırılır.


>Gerektiğinde ayrıntıları gizlemeye yardımcı olur.

**Önemli bulanıklaştırma türleri:**
>**Gaussian Blurring:** Gaussian blur, bir görüntünün Gaussian fonksiyonu ile bulanıklaştırılmasının sonucudur. Grafik yazılımlarında yaygın olarak kullanılan bir efekttir, genellikle görüntü gürültüsünü ve ayrıntıları azaltmak için kullanılır. Ayrıca makine öğrenimi veya derin öğrenme modellerimizi uygulamadan önce bir ön işleme aşaması olarak kullanılır. (kenarları yumuşatır)

>**Median Blur:** Median Filtresi, genellikle bir görüntü veya sinyalden gürültüyü gidermek için kullanılan doğrusal olmayan bir dijital filtreleme tekniğidir. Median filtreleme, belirli koşullar altında gürültüyü giderirken kenarları koruduğu için dijital görüntü işlemede çok yaygın olarak kullanılır. Tuz ve biber gürültüsünü gidermek için en iyi algoritmalardan biridir.

>**Bilateral Blur:** İkili filtre, görüntüler için doğrusal olmayan, kenar koruyan ve gürültü azaltan bir yumuşatma filtresidir. Her pikselin yoğunluğunu yakındaki piksellerden gelen yoğunluk değerlerinin ağırlıklı ortalamasıyla değiştirir. Bu ağırlık bir Gauss dağılımına dayanabilir. Böylece, keskin kenarlar korunurken zayıf olanlar atılır.

```python
import cv2 
import numpy as np 

image = cv2.imread('car2.jpg') 
image=cv2.resize(image,(400,500))
# Gaussian Blur 
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 

# Median Blur 
median = cv2.medianBlur(image, 5) 

# Bilateral Blur 
bilateral = cv2.bilateralFilter(image, 9, 75, 75) 

combined_image = cv2.hconcat([image, Gaussian, median, bilateral]) #çıktılarımızı birleştirmek


#cv2.imshow('Original Image', image) 
#cv2.imshow('Gaussian Blurring', Gaussian)
#cv2.imshow('Median Blurring', median) 
#cv2.imshow('Bilateral Blurring', bilateral) 
cv2.imshow('tumu', combined_image) 

#tümü olan görseli indirmek
cv2.imwrite("tum_blur.png", combined_image)

cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

# Görüntü Eşikleme(Threshold)

<code>cv2.threshold(source, thresholdValue, maxVal, thresholdingTechnique) </code>

-> **source:** Input Image array (must be in Grayscale). <br>
-> **thresholdValue:** Value of Threshold below and above which pixel values will change accordingly. <br>
-> **maxVal:** Maximum value that can be assigned to a pixel. <br>
-> **thresholdingTechnique:** The type of thresholding to be applied. <br>


### Eşikleme Mantığı Nedir?
Bir görüntüdeki her pikselin bir parlaklık veya yoğunluk değeri vardır. Eşikleme işlemi, bu piksel değerlerinin belirli bir eşik değerine göre karşılaştırılmasını sağlar. Bir pikselin değeri eşik değerinden küçükse veya büyükse, o piksel bir değere atanır (örneğin 0 ya da 255). Bu şekilde görüntüdeki belirli alanlar daha net bir şekilde ayırt edilir.

## Basit Eşikleme Teknikleri
Şimdi, OpenCV'deki bazı temel eşikleme yöntemlerini inceleyelim.

### 1. THRESH_BINARY:
Eşikleme işleminde en temel tekniktir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 255 (beyaz) olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, bu piksel değeri 0 (siyah) olarak atanır.<br>
Örnek: Bir eşik değeri 127 verilmişse, 128 ve üstü tüm pikseller beyaz yapılır, 126 ve altındaki tüm pikseller siyah yapılır.<br>

### 2. THRESH_BINARY_INV:
THRESH_BINARY'nin tam tersidir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 0 (siyah) olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, bu piksel değeri 255 (beyaz) olarak atanır.<br>
Örnek: 127 eşik değeriyle çalışıyorsak, 128 ve üstü pikseller siyah yapılırken, 126 ve altı tüm pikseller beyaz yapılır.<br>

### 3. THRESH_TRUNC:
Piksel değeri eşik değerinden büyükse, bu piksel değeri eşik değerine ayarlanır.<br>
Piksel değeri eşik değerinden küçükse, olduğu gibi bırakılır.<br>
Örnek: 127 eşik değeriyle, 130 değerinde olan bir piksel 127’ye çekilirken, 100 değerinde olan bir piksel aynı kalır.<br>

### 4. THRESH_TOZERO:
Piksel değeri eşik değerinden küçükse, bu piksel değeri 0 olarak atanır (siyah yapılır).<br>
Piksel değeri eşik değerinden büyükse, pikselin orijinal değeri korunur.<br>
Örnek: 127 eşik değeriyle, 126 ve altı değerlerdeki pikseller siyah yapılırken, 128 ve üstü değerler orijinal haliyle bırakılır.<br>

### 5. THRESH_TOZERO_INV:
THRESH_TOZERO işleminin tersidir.<br>
Piksel değeri eşik değerinden büyükse, bu piksel değeri 0 olarak atanır.<br>
Piksel değeri eşik değerinden küçükse, orijinal piksel değeri korunur.<br>
Örnek: 127 eşik değeriyle, 126 ve altı değerlerdeki pikseller orijinal haliyle bırakılırken, 128 ve üstü pikseller siyah yapılır.<br>

### Eşikleme Nerelerde Kullanılır?
Eşikleme genellikle nesne tespiti, kontur bulma veya nesne segmentasyonu gibi işlemlerde kullanılır. Örneğin, bir belge tarayıp sadece metinleri ön plana çıkarmak istediğinizde, eşikleme ile siyah-beyaz bir görüntü elde edebilirsiniz.

# Simple Thresholding

```python
import cv2  
import numpy as np  
  
# path to input image is specified and   
# image is loaded with imread command  
image = cv2.imread('apple1.jpg')  
  
# cv2.cvtColor is applied over the 
# image input with applied parameters 
# to convert the image in grayscale  
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
  
# applying different thresholding  
# techniques on the input image 
# all pixels value above 120 will  
# be set to 255 
ret, thresh1 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY) 
ret, thresh2 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY_INV) 
ret, thresh3 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TRUNC) 
ret, thresh4 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TOZERO) 
ret, thresh5 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_TOZERO_INV) 
  
# the window showing output images 
# with the corresponding thresholding  
# techniques applied to the input images 
cv2.imshow('Binary Threshold', thresh1) 
cv2.imshow('Binary Threshold Inverted', thresh2) 
cv2.imshow('Truncated Threshold', thresh3) 
cv2.imshow('Set to 0', thresh4) 
cv2.imshow('Set to 0 Inverted', thresh5) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Adaptive Thresholding
Basit Eşiklemede, her zaman sabit kalan global bir eşik değeri kullanılmıştır. Bu nedenle, sabit bir eşik değeri farklı alanlardaki değişken aydınlatma koşullarında yardımcı olmayacaktır. Uyarlanabilir eşikleme, eşik değerinin daha küçük bölgeler için hesaplandığı yöntemdir. Bu, aydınlatmadaki değişime göre farklı bölgeler için farklı eşik değerlerine yol açar. Bunun için cv2.adaptiveThreshold kullanıyoruz.

<code>cv2.adaptiveThreshold(source, maxVal, adaptiveMethod, thresholdType, blocksize, constant)</code>


**Parameters:**

-> **source:** Input Image array(Single-channel, 8-bit or floating-point)

-> **maxVal:** Maximum value that can be assigned to a pixel.

-> **adaptiveMethod:** Adaptive method decides how threshold value is calculated.


 **cv2.ADAPTIVE_THRESH_MEAN_C:** Threshold Value = (Mean of the neighbourhood area values – constant value). In other words, it is the mean of the blockSize×blockSize neighborhood of a point minus constant.


**cv2.ADAPTIVE_THRESH_GAUSSIAN_C:** Threshold Value = (Gaussian-weighted sum of the neighbourhood values – constant value). In other words, it is a weighted sum of the blockSize×blockSize neighborhood of a point minus constant.

-> **thresholdType:** The type of thresholding to be applied.

-> **blockSize:** Size of a pixel neighborhood that is used to calculate a threshold value.

-> **constant:** A constant value that is subtracted from the mean or weighted sum of the neighbourhood pixels.

```python
import cv2  
import numpy as np  
   
# path to input image is specified and   
# image is loaded with imread command  
image = cv2.imread('apple1.jpg')  
   
# cv2.cvtColor is applied over the 
# image input with applied parameters 
# to convert the image in grayscale  
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
   
# applying different thresholding  
# techniques on the input image 
thresh1 = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
  
thresh2 = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
  
# the window showing output images 
# with the corresponding thresholding  
# techniques applied to the input image 
cv2.imshow('Adaptive Mean', thresh1) 
cv2.imshow('Adaptive Gaussian', thresh2) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Morfolojik Operasyonlar

## Erosion(aşındırma) ve Dilation(genişletme)

```python
import cv2 
import numpy as np 
  
# Reading the input image 
img = cv2.imread('ataturk.png')
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

thresh,BW_im = cv2.threshold(gray_img,150,255,cv2.THRESH_BINARY)


  
# Taking a matrix of size 5 as the kernel 
kernel = np.ones((5,5), np.uint8) 
  
# The first parameter is the original image, 
# kernel is the matrix with which image is 
# convolved and third parameter is the number 
# of iterations, which will determine how much 
# you want to erode/dilate a given image. 
img_erosion = cv2.erode(BW_im, kernel, iterations=1) 
img_dilated=cv2.dilate(BW_im, kernel, iterations=1) 

  
cv2.imshow('orijinal Gorsel', img) 
cv2.imshow('siyah-beyaz', BW_im) 
cv2.imshow('Erozyon uygulanmis', img_erosion) 
cv2.imshow('Dilasyon uygulanmis', img_dilated) 

cv2.imwrite("erozyonlu.png",img_erosion)
cv2.imwrite("dilasyonlu.png",img_dilated)





  
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Opening

Opening, gürültüyü temizlemek ve istenmeyen küçük nesneleri kaldırmak için idealdir. Özellikle, nesnenin genel yapısını koruyarak küçük beyaz piksellerden (gürültüden) kurtulmak için kullanılır.

<code>cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)</code>

**Parameters:** <br>
-> **image:** Input Image array. <br>
-> **cv2.MORPH_OPEN:** Applying the Morphological Opening operation. <br>
-> **kernel:** Structuring element. 

```python
import numpy as np
import cv2

img = cv2.imread("apple1.jpg",0)

ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)

kernel = np.ones((7,7),dtype=np.uint8)

whiteNoise = np.random.randint(0,2,size=thresh1.shape[:2])
whiteNoise = whiteNoise*255
noise_img = whiteNoise + thresh1

opening = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_OPEN,kernel)

cv2.imshow("Original",img)
cv2.imshow("threshold",thresh1)
cv2.imshow("Opening",opening)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Closing

**Kapanma (Closing)** işlemi OpenCV'de kullanılan bir morfolojik operasyondur ve temelde bir görüntüdeki küçük boşlukları veya ince siyah alanları kapatmak için kullanılır. Kapanma, bir görüntüye ilk olarak bir genişletme (dilation) ve ardından bir aşındırma (erosion) işlemi uygulayarak gerçekleştirilir.<br>

Genişletme adımında beyaz alanlar genişletilir ve küçük siyah boşluklar kapanır. Ardından yapılan aşındırma işlemi, genişletme sonrası görüntünün boyutlarını orijinal şekline yaklaştırır. Bu yöntem, genellikle nesne çevresindeki küçük deliklerin veya nesneler arasındaki ince boşlukların doldurulmasında etkilidir.

<code>cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)</code><br>


**Parameters:**<br>

-> **image:** Input Image array.<br>

-> **cv2.MORPH_CLOSE:** Applying the Morphological Closing operation.<br>

-> **kernel:** Structuring element.<br>

>Haar Cascade sınıflandırıcısı, bilgisayarlı görü çalışmalarında nesne tespiti için kullanılan bir makine öğrenimi algoritmasıdır. Yani, bir resimde belirli nesneleri bulmak için kullanılır. Örneğin, yüzler, gözler, araçlar gibi nesnelerin tespitinde yaygın olarak kullanılır.En temel manada belirli bir algoritmaya göre bulunması istenen nesneler önce bilgisayara tanıtılır ve daha sonra ona benzer şekillerin bulunduğu resimler veya video frameleri taranarak o nesne bulunmaya çalışılır.

```python
import cv2

cascade_path=cv2.data.haarcascades+"haarcascade_frontalface_default.xml"
"haarcascade_eye.xml"
eye_path=cv2.data.haarcascades+"haarcascade_eye.xml"

face_cascaded=cv2.CascadeClassifier(cascade_path)
eye_cascade=cv2.CascadeClassifier(eye_path)


img_path="real2.jpeg"

img=cv2.imread(img_path)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces=face_cascaded.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30,30))

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

    roi_gray=gray[y:y+h,x:x+w]
    roi_color=img[y:y+h,x:x+w]
    eyes=eye_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=2,minSize=(30,30))

    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ew+ex,eh+ey),(0,255,0),2)

cv2.imshow("pencere",img)
cv2.imwrite("haarcascade2.png",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Kontür Bulma Ve Çizme

**Konturlar**, bir görüntünün sınırı boyunca aynı yoğunluğa sahip tüm noktaları birleştiren çizgi olarak tanımlanır. Konturlar şekil analizinde, ilgilenilen nesnenin boyutunu bulmada ve nesne tespitinde kullanışlıdır.
OpenCV, görüntüden konturların çıkarılmasına yardımcı olan <code>findContour()</code> işlevine sahiptir. <br>
En iyi ikili(binary) görüntülerde çalışır, bu nedenle önce eşikleme teknikleri, Sobel kenarları vb. uygulamalıyız.



```python
import cv2 
import numpy as np 

image = cv2.imread('apple1.jpg',0)

thresh = cv2.adaptiveThreshold(
    image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print("Number of Contours found = " + str(len(contours))) 

cv2.drawContours(image, contours, -1, (0, 255, 0), 3) 

cv2.imshow("elma",image)
cv2.imshow("thresholding",thresh)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Kontür Sayma ve şekil çizdirme

```python
import cv2
import numpy as np

# 1. Resmi okuma ve gri tonlamalı hale getirme
image = cv2.imread('elma.jpg') 
image_yan=cv2.imread('elma_yandan.jpg')
image=cv2.resize(image,(900,900))
image_yan=cv2.resize(image_yan,(900,900))


#grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_yan = cv2.cvtColor(image_yan, cv2.COLOR_BGR2GRAY)

# 2. Gürültü azaltma (Gaussian Blur)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
blurred_yan = cv2.GaussianBlur(gray_yan, (5, 5), 0)
# 3. Adaptive Threshold ile siyah-beyaza çevirme


thresh = cv2.adaptiveThreshold(
    blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

thresh_yan= cv2.adaptiveThreshold(
    blurred_yan, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

# 4. Konturları bulma
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours_yan, _ = cv2.findContours(thresh_yan, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


# En büyük konturu bulma (Elma konturunu seçme)
contour = max(contours, key=cv2.contourArea)
contour_yan = max(contours_yan, key=cv2.contourArea)

# 5. En uygun çemberi bulma (minEnclosingCircle)
(x, y), radius = cv2.minEnclosingCircle(contour)
center = (int(x), int(y))
radius = int(radius)

(x2, y2), radius_yan= cv2.minEnclosingCircle(contour_yan)
center_yan = (int(x2), int(y2))
radius_yan = int(radius_yan)

# Çemberin alanını hesaplama
circle_area = np.pi * (radius ** 2)
yukseklik=radius_yan*2

# Çemberi ve konturları çizme
output_image = image.copy()
cv2.circle(output_image, center, radius, (0, 255, 0), 2)  # Yeşil renkli çember
cv2.drawContours(output_image, [contour], -1, (255, 0, 0), 2)  # Mavi renkli kontur

#yandan görüntüye yapılan işlemler
output_image_yan= thresh_yan.copy()
cv2.circle(output_image_yan, center_yan, radius_yan, (0,255,0), 2)
print(center_yan)

start_point = (center_yan[0], center_yan[1] - radius_yan)  # Üst nokta
end_point = (center_yan[0], center_yan[1] + radius_yan)    # Alt nokta
cv2.line(output_image_yan, start_point, end_point, (255, 255, 0), 2)

text3 = f" {radius_yan*2:.2f}"
cv2.putText(
    output_image_yan,  # Görüntü
    text3,  # Yazılacak metin
    (150, 161),  # Metnin başlangıç noktası (x, y)
    cv2.FONT_HERSHEY_SIMPLEX,  # Yazı tipi
    1,  # Yazı boyutu (fontScale) artırıldı
    (0,0,0),  # Yazı rengi (Kırmızı)
    2,  # Yazı kalınlığı artırıldı
    cv2.LINE_AA  # Yazı çizim tipi
)
volume=1.33*circle_area*yukseklik
# 6. Çember hacmini görüntüye yazdırma
text2 = f"Hacim: {volume:.2f} px^3"
cv2.putText(
    output_image_yan,  # Görüntü
    text2,  # Yazılacak metin
    (10, 30),  # Metnin başlangıç noktası (x, y)
    cv2.FONT_HERSHEY_SIMPLEX,  # Yazı tipi
    1,  # Yazı boyutu (fontScale) artırıldı
    (0, 0, 255),  # Yazı rengi (Kırmızı)
    2,  # Yazı kalınlığı artırıldı
    cv2.LINE_AA  # Yazı çizim tipi
)


# 6. Çember alanını görüntüye yazdırma
text = f"Alan: {circle_area:.2f} px^2"
cv2.putText(
    output_image,  # Görüntü
    text,  # Yazılacak metin
    (10, 30),  # Metnin başlangıç noktası (x, y)
    cv2.FONT_HERSHEY_SIMPLEX,  # Yazı tipi
    1,  # Yazı boyutu (fontScale) artırıldı
    (0, 0, 255),  # Yazı rengi (Kırmızı)
    2,  # Yazı kalınlığı artırıldı
    cv2.LINE_AA  # Yazı çizim tipi
)
print("yuk: ",yukseklik)

# 7. Sonuçları gösterme
cv2.imshow('Original Image', image)
cv2.imshow('Original_yan Image', image_yan)
cv2.imshow('Threshold Image', thresh)
cv2.imshow('Threshold_yan image', thresh_yan)
cv2.imshow('Contour and Circle', output_image)
cv2.imshow('yandan', output_image_yan)

print(f"Çemberin Alanı: {circle_area:.2f} birim kare")
print(f"Çemberin hacmi: {volume:.2f} birim küp")

cv2.imwrite("alan.png",output_image)
cv2.imwrite("adaptive_thresholding.png",thresh)
cv2.imwrite("hacim.png",output_image_yan)
cv2.imwrite("adaptive_thresholding_yan.png",thresh_yan)
cv2.imwrite("yan_thth.png",output_image_yan)



cv2.waitKey(0)
cv2.destroyAllWindows()
```

# çember çizme

```python
import cv2
import numpy as np
import math

# Başlangıç noktası ve çizim durumu için değişkenler
start_point = None
drawing = False  # Çizim işlemi devam ediyor mu?

# Fare olayını yakalayan fonksiyon
def mouse_callback(event, x, y, flags, param):
    global start_point, drawing, img, temp_img

    # Sol fare tuşuna basıldığında çizimi başlat
    if event == cv2.EVENT_LBUTTONDOWN:
        start_point = (x, y)
        drawing = True

    # Fare hareket ederken geçici bir çember çiz
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:  # Sadece sol tıklama basılıysa
            temp_img = img.copy()  # Geçici çizim için görüntüyü kopyala
            radius = int(math.sqrt((x - start_point[0]) ** 2 + (y - start_point[1]) ** 2))
            cv2.circle(temp_img, start_point, radius, (0, 255, 0), 2)
            cv2.imshow("Dynamic Circle", temp_img)

    # Sol fare tuşu bırakıldığında çemberi son haliyle çiz
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        radius = int(math.sqrt((x - start_point[0]) ** 2 + (y - start_point[1]) ** 2))
        cv2.circle(img, start_point, radius, (0, 255, 0), 2)
        cv2.imshow("Dynamic Circle", img)

# Boş bir görüntü oluştur
img = np.zeros((512, 512, 3), np.uint8)
temp_img = img.copy()  # Geçici çizimler için kopya görüntü

# Pencere oluştur ve fare geri çağırma fonksiyonunu ayarla
cv2.namedWindow("Dynamic Circle")
cv2.setMouseCallback("Dynamic Circle", mouse_callback)

while True:
    # Ana görüntüyü göster
    cv2.imshow("Dynamic Circle", img)
    # 'q' tuşuna basıldığında çık
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
